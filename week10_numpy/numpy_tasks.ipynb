```python
# Task 31: Manual min-max normalization for a 2D matrix column-wise
m = np.random.randint(1, 100, (4, 3)).astype(float)
(m - m.min(axis=0)) / (m.max(axis=0) - m.min(axis=0))
```

```python
# Task 32: Compute pairwise Euclidean distances between rows of a matrix
def pairwise_distances(X):
    diff = X[:, np.newaxis, :] - X[np.newaxis, :, :]
    return np.sqrt((diff ** 2).sum(axis=2))

m = np.random.rand(4, 3)
pairwise_distances(m)
```

```python
# Task 33: Compute the covariance matrix manually using NumPy
m = np.random.randint(1, 100, (5, 4)).astype(float)
centered = m - m.mean(axis=0)
cov = (centered.T @ centered) / (m.shape[0] - 1)
cov
```

```python
# Task 35: Implement batch gradient descent for linear regression using only NumPy
np.random.seed(42)
X = np.random.rand(100, 1)
y = 3 * X.squeeze() + 2 + np.random.randn(100) * 0.1

X_b = np.c_[np.ones(100), X]  # add bias column
theta = np.zeros(2)
lr, epochs = 0.1, 1000

for _ in range(epochs):
    grad = X_b.T @ (X_b @ theta - y) / len(y)
    theta -= lr * grad

print(f"intercept: {theta[0]:.4f}, slope: {theta[1]:.4f}")
```

```python
# Task 36: Compute eigenvalues and eigenvectors and verify the result
m = np.array([[4, 2], [1, 3]], dtype=float)
eigenvalues, eigenvectors = np.linalg.eig(m)
print("eigenvalues:", eigenvalues)
# Verify: A @ v == lambda * v for each eigenvector
for i in range(len(eigenvalues)):
    v = eigenvectors[:, i]
    print(f"verified λ{i}: {np.allclose(m @ v, eigenvalues[i] * v)}")
```

```python
# Task 37: Implement softmax function for a 2D array (row-wise) with numerical stability
def softmax(X):
    X_stable = X - X.max(axis=1, keepdims=True)
    exp = np.exp(X_stable)
    return exp / exp.sum(axis=1, keepdims=True)

m = np.random.randn(3, 4)
softmax(m)
```

```python
# Task 38: Perform SVD on a matrix and reconstruct it
m = np.random.randint(1, 10, (4, 3)).astype(float)
U, S, Vt = np.linalg.svd(m, full_matrices=False)
reconstructed = U @ np.diag(S) @ Vt
print("original:\n", m)
print("reconstructed:\n", np.round(reconstructed, 6))
```

```python
# Task 39: Implement k-means clustering (k=3) using only NumPy
np.random.seed(0)
X = np.vstack([np.random.randn(50, 2) + c for c in [(0, 0), (5, 5), (0, 5)]])
k, epochs = 3, 100

centroids = X[np.random.choice(len(X), k, replace=False)]
for _ in range(epochs):
    dists = np.linalg.norm(X[:, np.newaxis] - centroids, axis=2)
    labels = dists.argmin(axis=1)
    centroids = np.array([X[labels == i].mean(axis=0) for i in range(k)])

print("centroids:\n", centroids)
```

```python
# Task 40: Neural network forward pass (1 hidden layer) using only NumPy
np.random.seed(0)
def relu(x): return np.maximum(0, x)

# Input: 4 samples, 3 features; hidden: 4 neurons; output: 2 neurons
X = np.random.randn(4, 3)
W1 = np.random.randn(3, 4) * 0.1
b1 = np.zeros(4)
W2 = np.random.randn(4, 2) * 0.1
b2 = np.zeros(2)

hidden = relu(X @ W1 + b1)
output = softmax(hidden @ W2 + b2)
print("output probabilities:\n", np.round(output, 4))
```
