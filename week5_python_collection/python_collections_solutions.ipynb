```python
def flatten(nested_list):
    result = []
    for item in nested_list:
        if type(item) == list:
            result = result + flatten(item)
        else:
            result.append(item)
    return result

nested = [1, [2, 3], [4, [5, 6]], 7]
print(flatten(nested))
```

```python
coordinates = {}
coordinates[(0, 0)] = "origin"
coordinates[(3, 4)] = "point A"
coordinates[(6, 8)] = "point B"

point1 = (3, 4)
point2 = (6, 8)
manhattan_distance = abs(point2[0] - point1[0]) + abs(point2[1] - point1[1])
print(manhattan_distance)
```

```python
list_of_lists = [[1, 2, 3], [2, 3, 4], [3, 4, 5]]
all_unique = set()
for lst in list_of_lists:
    for item in lst:
        all_unique.add(item)
print(all_unique)
```

```python
dict1 = {'a': 1, 'b': 2, 'c': 3}
dict2 = {'b': 3, 'c': 4, 'd': 5}
merged = {}
for key in dict1:
    merged[key] = dict1[key]
for key in dict2:
    if key in merged:
        merged[key] = merged[key] + dict2[key]
    else:
        merged[key] = dict2[key]
print(merged)
```
